# Implement a wordcount program using map, reduce and sorting

```python
import re
from itertools import groupby
from operator import itemgetter
from functools import reduce
```

```python
def tokenize(text: str):
    return re.findall(r"[a-z0-9]+(?:'[a-z0-9]+)?", text.lower())
```

```python
def wordcount_map_reduce_sort(text: str, sort_by="count_desc"):
    # Option 1
    words = tokenize(text)

    pairs = list(map(lambda w: (w, 1), words))

    pairs.sort(key=lambda x: x[0])

    counts = [
        (word, reduce(lambda acc, pair: acc + pair[1], group, 0))
        for word, group in groupby(pairs, key=lambda x: x[0])
    ]

    ###########
    #  Option 2
    # mapped = map(lambda w: (w, 1), tokenize(text))

    # mapped_sorted_by_word = sorted(mapped, key=itemgetter(0))

    # reduced = (
    #     (word, sum(count for _, count in group))
    #     for word, group in groupby(mapped_sorted_by_word, key=itemgetter(0))
    # )

    # counts = list(reduced)
    ###########
    if sort_by == "count_desc":
        counts.sort(key=lambda x: (-x[1], x[0]))
    elif sort_by == "count_asc":
        counts.sort(key=lambda x: (x[1], x[0]))
    elif sort_by == "alphabetic":
        counts.sort(key=lambda x: x[0])
    else:
        raise ValueError("sort_by must be one of: count_desc, count_asc, alphabetic")

    return counts
```

```python
sample = """Lorem Ipsum is simply dummy text of the printing and typesetting industry.
Lorem Ipsum has been the industry's standard dummy text ever since the 1500s, when an
unknown printer took a galley of type and scrambled it to make a type specimen book."""
print(wordcount_map_reduce_sort(sample, sort_by="count_asc"))
```
